```python
from langchain_community.document_loaders import TextLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma
```

```python
loaded_document = TextLoader('d:/lecture/langchainlive/data/state_of_the_union.txt').load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
```

```python
chunks_of_text = text_splitter.split_documents(loaded_document)
```

```python
chunks_of_text
```

```python
len(chunks_of_text)
```

```python
vector_db = Chroma.from_documents(chunks_of_text, HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2"))
```

```python
vector_db
```

```python
question = "What did the president say about the John Lewis Voting Rights Act?"
```

```python
response = vector_db.similarity_search(question)
```

```python
print(response[0].page_content)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("d:/lecture/langchainlive/data/state_of_the_union.txt")
```

```python
pip install faiss-cpu
```

```python
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loaded_document = loader.load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
```

```python
chunks_of_text = text_splitter.split_documents(loaded_document)
```

```python
embedding = HuggingFaceBgeEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector_db = FAISS.from_documents(chunks_of_text, embedding)
```

```python
vector_db 
```

```python
retriever = vector_db.as_retriever(search_kwargs={"k":3})
```

```python
response=retriever.invoke("what did he say about ketanji brown jackson?")
```
